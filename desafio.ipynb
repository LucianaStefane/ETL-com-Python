# Transformação

```python
df = df.rename(columns={
    'PassengerId': 'id',
    'Survived': 'sobreviveu',
    'Pclass': 'classe',
    'Name': 'nome', 
    'Sex': 'sexo',
    'Age': 'idade',
    'SibSp': 'irmaos_abordo',
    'Parch': 'pais_abordo',
    'Ticket': 'bilhete',
    'Fare': 'tarifa',
    'Cabin': 'cabine',
    'Embarked': 'cidade_embarque'
})
```

```python
df['classe'] = df['classe'].map({1: 'A', 2: 'B', 3: 'C'})
df['sobreviveu'] = df['sobreviveu'].map({0: 'não sobreviveu', 1: 'sobreviveu'})
df['cidade_embarque'] = df['cidade_embarque'].map({'C': 'Cherbourg', 'Q': 'Queenstown', 'S': 'Southampton'})
df['irmaos_abordo'] = df['irmaos_abordo'].map({0: 'Não acompanhado', 1: 'Acompanhado'})
df['pais_abordo'] = df['pais_abordo'].map({0: 'Não acompanhado', 1: 'Acompanhado'})
df['tratamento'] = df['sexo'].map({'female': 'Sra.', 'male': 'Sr.'})
```

```python
df['nome_completo'] = df['nome'].str.split(pat=',', expand=True)[1].str.split(pat='.', expand=True)[1] + ' ' + df['nome'].str.split(pat=',', expand=True)[0]
```

```python
df['idade'] = df['idade'].fillna(-1).astype('int').astype('string').str.replace('-1', 'desconhecidos')
df['cidade_embarque'] = df['cidade_embarque'].fillna('cidade desconhecida')
df['cabine'] = df['cabine'].fillna('desconhecida')
df[['irmaos_abordo', 'pais_abordo']] = df[['irmaos_abordo', 'pais_abordo']].fillna('Não acompanhado')
df['tarifa'] = df['tarifa'].round(2)
```

```python
df_final = pd.DataFrame()
df_final['frase'] = df.apply(lambda row: f'O/A {row["tratamento"]} {row["nome_completo"]} {row["sobreviveu"]} ao acidente do Titanic. Na época, contava com {row["idade"]} anos, embarcou em {row["cidade_embarque"]}, na cabine {row["cabine"]}, classe {row["classe"]}. Pagou US$ {row["tarifa"]} pelo bilhete {row["bilhete"]}. Embarcou {row["irmaos_abordo"]} de irmãos. Embarcou {row["pais_abordo"]} de pais.', axis=1)
```

# Carregamento

```python
df_final.to_csv('./dados/result/status.csv')
```
